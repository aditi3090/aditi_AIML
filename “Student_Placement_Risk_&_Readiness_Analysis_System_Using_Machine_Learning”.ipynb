Salary Prediction (Linear Regression)

```python
# take only placed students
df_salary = df[df['status'] == 1]

X_sal = df_salary.drop(['status', 'salary'], axis=1)
y_sal = df_salary['salary']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_sal, y_sal, test_size=0.2, random_state=42)

model_sal = LinearRegression()
model_sal.fit(X_train_s, y_train_s)

y_pred_s = model_sal.predict(X_test_s)

from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test_s, y_pred_s))
```

```python
df['readiness_score'] = (
    0.4 * df['overall_score'] +
    0.3 * df['etest_p'] +
    0.3 * (df['workex'] * 100)
)
```

```python
def readiness_level(score):
    if score >= 75:
        return "High"
    elif score >= 60:
        return "Medium"
    else:
        return "Low"

df['readiness_level'] = df['readiness_score'].apply(readiness_level)
```

```python
df = pd.get_dummies(df, columns=['readiness_level'], drop_first=True)
```

```python
def risk_level(score):
    if score < 60:
        return "High Risk"
    elif score < 75:
        return "Medium Risk"
    else:
        return "Low Risk"

df['risk_level'] = df['readiness_score'].apply(risk_level)
```

```python
df = pd.get_dummies(df, columns=['risk_level'], drop_first=True)
```

```python
def student_analysis(pred, readiness, risk):

    # placement message
    if pred == 1:
        placement = "High Probability of Placement"
    else:
        placement = "Low Probability of Placement"

    # readiness message + motivation
    if readiness == "High":
        readiness_msg = "You are well-prepared and ready to grab opportunities."
        motivation = "Keep pushing forward — success is within your reach!"
    elif readiness == "Medium":
        readiness_msg = "You are on the right track but there is room for improvement."
        motivation = "Stay consistent and keep improving — you’re closer than you think!"
    else:
        readiness_msg = "You need to work on your skills and preparation."
        motivation = "Don’t give up — every expert was once a beginner. Start improving today!"

    # risk message
    if risk == "Low Risk":
        risk_msg = "Low risk — your profile is strong."
    elif risk == "Medium Risk":
        risk_msg = "Moderate risk — focus on improving key areas."
    else:
        risk_msg = "High risk — immediate improvement is required to boost your chances."

    # final output
    return f"""
🔹 Placement Prediction: {placement}
🔹 Readiness Level: {readiness}
🔹 Risk Level: {risk}

📊 Analysis:
{readiness_msg}
{risk_msg}

💡 Motivation:
{motivation}
"""
```
